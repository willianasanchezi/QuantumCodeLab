### Librerias

```python
import qiskit

from qiskit import QuantumCircuit # Librerias para definir los circuitos cuanticos
from qiskit_aer import Aer
from qiskit.primitives import Sampler
from qiskit.visualization import plot_bloch_multivector
from mpl_toolkits import mplot3d
```

### Definir un circuito, de un solo qbit

```python
qc = QuantumCircuit(1)
```

### Inicializar qbit con valor 1

```python
initial_state = [0, 1]

qc.initialize(initial_state, 0)
```

### Obtener el backend del simulador de estado

```python
backend = Aer.get_backend('statevector_simulator')
```

### Ejecutar el circuito usando backend.run()

```python
job = backend.run(qc)
```

```python
statevector = job.result().get_statevector()
print(statevector)
```

### Graficar el vector de estado en el Bloch sphere usando plot_bloch_multivector

```python
plot_bloch_multivector(statevector)
```

### Aplicando Puerta Not X

```python
qc.x(0)
```

```python
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc)
statevector = job.result().get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
qc.h(0)
```

```python
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc)
statevector = job.result().get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```
